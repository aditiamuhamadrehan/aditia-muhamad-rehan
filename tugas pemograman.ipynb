```python
import numpy as np

array1 = np.array([1, 2, 3])
array2 = np.array([4, 5, 6])

hasil = np.add(array1, array2)

hasil_operator = array1 + array2

print("Hasil penjumlahan menggunakan np.add:", hasil)
print("Hasil penjumlahan menggunakan operator +:", hasil_operator)
```

```python
import pandas as pd

data = {
    'A': [1, 2, 3],
    'B': [5, 4, 6]
}
df = pd.DataFrame(data)

print("DataFrame awal:")
print(df)

df['C'] = df['A'] + df['B']

df['D'] = df['A'] - df['B']

df['E'] = df['A'] * df['B']

df['F'] = df['A'] / df['B']

print("\nDataFrame setelah operasi aritmatika:")
print(df)
```

```python
import pandas as pd
import numpy as np

np.random.seed(42)

data = {
    'A': np.random.randint(1, 100, size=10),
    'B': np.random.rand(10) * 100
}
df = pd.DataFrame(data)

print("DataFrame:")
print(df)

print("\n5 data teratas:")
print(df.head())

print("\n5 data terbawah:")
print(df.tail())
```
